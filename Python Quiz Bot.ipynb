```python

# Functions for Creating an instance of a class -- an instance of a type of object, a particular detail of data


### IMMUTABLE TYPES 


### these functions create a random instance of str;int;float;type respectively
def make_str(max_length=8):
    """make a random assortment of characters """
    calculated_length = random.randrange(1,max_length)
    characters = [x for x in string_functions.ascii_lowercase]
    # * 2 + [x for x in """<>?/"':;|\}]{[+=_-*%)($%^&#@!~`"""]
    potential_words = ['False','def','if','raise','None','del','import','return','True','elif','in','try','and','else','is','while',
                              'as','except','lambda','with','assert','finally','nonlocal',
                              'yield','break','for','not','class','form','or','continue','global','pass'] + [''.join([random.choice(characters) for i in range(calculated_length)]).capitalize() for x in range(50)]
    output = random.choice(potential_words)
#         if output in self.reserved_names:
#             print('thats a reserved name in Python')
    return output

def make_int(scale=50):
    return int(numpy.random.normal(0,scale))

def make_float(scale=50):
    return random.choice([round(x + y,2) for x, y in zip([float(x) for x in numpy.random.normal(0,scale,size=100)], [int(x) for x in numpy.random.random_sample(100)])])

def make_type(types=[str, list, int, dict, set, tuple, float, type]):
    return random.choice(types) # note that we can't actually include `class` in this list, nor to call type() on it

def make_a_value(kinds=['all']):  
    '''links to the functions above, as well as any others we might make'''
    generators = [make_str, make_int, make_float, make_type, make_tuple, make_list]
    types = [str, int, float,tuple, list]
    mapping = dict(zip(types,generators))

    if 'all' in kinds:
        kind = random.choice(types)
    else:
        kind = random.choice(kinds)
    
    
    # call the mapped function
    value = mapping[kind]()

    if isinstance(value, (list,tuple,set)):
        while len(value) > 3:
            for x in value:
                if isinstance(x, (list,tuple,set)):
                    if len(x) > 2:
                        value = mapping[kind]()

    if type(value) == dict:
        while len(value) > 2:
            for x in value.values():
                if isinstance(x, (list,tuple,set)):
                    if len(x) > 2:
                        value = mapping[kind]()
    return value

#def sample_values(size=100):
    

#all_possible = singular_value_sample()


#### DATA CONTAINERS



def make_tuple(size=None):
    if not size:
        length = int(numpy.random.normal(3, .667))
    else:
        length = size
    output = tuple([make_a_value() for x in range(length)])
    return output

def make_list():
    length = int(numpy.random.normal(3, .667))
    output = [make_a_value(kind=[str,int,float,tuple]) for x in range(length)]
    #types = [type(x) for x in outputs]
    return output


#     def make_tuple(length=4)
def make_key():
    random.randint(0,9) + make_str(max_length=2)
    
    
#def make_dict():
    
```
